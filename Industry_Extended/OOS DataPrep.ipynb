```python
industry_daily = pd.read_csv('C:/Users/95att/Desktop/job/First_paper_QAC/Volatility Management/10_Industry_Portfolios_Daily.csv')
```

```python
industry_daily.set_index('Unnamed: 0',inplace=True)
industry_daily.index.names = ['Date']
industry_daily = industry_daily.iloc[:26129].copy()

df = industry_daily.copy()
df.index = pd.to_datetime(df.index, format='%Y%m%d')
```

```python
df.tail(), df.head()
```

```python
start_day = '2013-12-14'

df_long_daily = (
    df.rename_axis('date').reset_index()
      .melt(id_vars='date', var_name='tic', value_name='close')
      .dropna(subset=['close'])
      .sort_values(['tic','date']).reset_index(drop=True)
)
df_long_daily.head()
```

```python
df_long_daily = df_long_daily[df_long_daily['date'] >= start_day]
df_long_daily
```

```python
df_long_daily.to_csv('Industry_long_daily_extended.csv', index=False)
```
